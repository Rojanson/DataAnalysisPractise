```python
def generate_circles(number):
    x_data = list()
    y_data = list()
    while len(x_data) < number:
        X = np.random.uniform(-2, 2, 100)
        Y = np.random.uniform(-2, 2, 100)
        mask = ((X - 1)**2 + (Y - 1)**2 <= 0.5) | ((X + 1)**2 + (Y - 1)**2 <= 0.5) | ((X - 1)**2 + (Y + 1)**2 <= 0.5) | ((X + 1)**2 + (Y + 1)**2 <= 0.5)
        
        x_data.append(X[mask][0])
        y_data.append(Y[mask][0])
    x_data = np.array(x_data)
    y_data = np.array(y_data)
    return x_data, y_data
```

```python
x, y = generate_circles(2000)
plt.scatter(x, y)
```

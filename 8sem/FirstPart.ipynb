```python
res = sklearn.cluster.DBSCAN(eps=0.058, min_samples=6).fit(data[:, :2])
len(np.unique(res.labels_))
```

```python
optics_min = [9, 7, 6]
dbscan_eps = [0.055, 0.056, 0.058]

for i, n in enumerate([2, 3, 5]):
    fig, ax = plt.subplots(1, 4, figsize=(12, 3))
    cluster0 = sklearn.cluster.KMeans(n_clusters=n).fit(data[:, :2])
    cluster1 = sklearn.cluster.DBSCAN(eps=dbscan_eps[i], min_samples=6).fit(data[:, :2])
    cluster2 = sklearn.cluster.OPTICS(eps=5, min_samples=optics_min[i]).fit(data[:, :2])
    cluster3 = sklearn.cluster.SpectralClustering(n_clusters=n).fit(data[:, :2])

    for k in range(len(data)):
        ax[0].scatter(data[k, 0], data[k, 1], color=colors[int(cluster0.labels_[k])])
    ax[0].set_title('KMeans')   
    for k in range(len(data)):
        ax[1].scatter(data[k, 0], data[k, 1], color=colors[int(cluster1.labels_[k])])
    ax[1].set_title('DBSCAN') 
    for k in range(len(data)):
        ax[2].scatter(data[k, 0], data[k, 1], color=colors[int(cluster2.labels_[k])])
    ax[2].set_title('OPTICS')    
    for k in range(len(data)):
        ax[3].scatter(data[k, 0], data[k, 1], color=colors[int(cluster3.labels_[k])])
    ax[3].set_title('SpectralClustering')
```

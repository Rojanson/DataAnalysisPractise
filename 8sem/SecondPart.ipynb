```python
import numpy as np
import pandas as pd
import sklearn.cluster
```

```python
from scipy.cluster import hierarchy

import matplotlib.pyplot as plt
```

```python
ytdist = data

Z = hierarchy.linkage(ytdist, 'single')

plt.figure(figsize=(16, 5))

dn = hierarchy.dendrogram(Z)
```

```python
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df.head()
```

```python
data = np.array(df)
```

```python
data[data == 'Iris-setosa'] = 0
data[data == 'Iris-versicolor'] = 1
data[data == 'Iris-virginica'] = 2
```

```python
ytdist = data[:, :4]

Z = hierarchy.linkage(ytdist, 'single')

plt.figure(figsize=(16, 5))

dn = hierarchy.dendrogram(Z)
```

```python
cluster0 = sklearn.cluster.KMeans(n_clusters=2).fit(ytdist)
cluster1 = sklearn.cluster.SpectralClustering(n_clusters=2).fit(ytdist)
```

```python
for i in range(len(data)):
    plt.scatter(data[i, 0], data[i, 1], color=colors[int(data[i, 4])])
```

```python
for i in range(len(data)):
    plt.scatter(data[i, 0], data[i, 1], color=colors[int(cluster0.labels_[i])])
```

```python
for i in range(len(data)):
    plt.scatter(data[i, 0], data[i, 1], color=colors[int(cluster1.labels_[i])])
```

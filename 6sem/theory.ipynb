```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import math
from sklearn import linear_model
import sklearn
```

```python
eps = np.random.standard_t(3, 100)

x = np.random.uniform(0, 1, 100)

y = 0.1*x**2 + 3*x + 5 + eps

plt.scatter(x, y)
```

```python
Liner = sklearn.linear_model.QuantileRegressor(quantile=0.5, alpha=0.3).fit(x[:, None], y)

print('Parameters')
print('sec:', Liner.score(x[:, None], y))
print('Коэффициенты модели:', Liner.coef_)
print('Свободный член:', Liner.intercept_)

a = np.array(Liner.coef_)
b = Liner.intercept_

plt.scatter(x, a*x + b)
plt.scatter(x, y)
```

```python
Liner = sklearn.linear_model.HuberRegressor(epsilon=2).fit(x[:, None], y)

print('Parameters')
print('sec:', Liner.score(x[:, None], y))
print('Коэффициенты модели:', Liner.coef_)
print('Свободный член:', Liner.intercept_)

a = np.array(Liner.coef_)
b = Liner.intercept_

plt.scatter(x, a*x + b)
plt.scatter(x, y)
```

```python
Liner = sklearn.linear_model.RANSACRegressor(min_samples=40, max_trials=20).fit(x[:, None], y)


plt.scatter(x, Liner.predict(x[:, None]))
plt.scatter(x, y)
```

```python
import statsmodels.api as sm
```

```python
n = 100
X = np.random.uniform(1, 1.4, (n, 1))
```

```python
a = [4]

teta = X @ a

Y = np.random.poisson(np.exp(teta), n)
```

```python
poisson_model = sm.GLM(Y, X, family=sm.families.Poisson())
```

```python
fit_result = poisson_model.fit()
```

```python
fit_result.summary()
```

```python
Liner = sklearn.linear_model.PoissonRegressor().fit(X, Y)
```

```python
mus = np.exp(X@Liner.coef_ + Liner.intercept_)
```

```python
Liner.coef_, Liner.intercept_
```

```python
((Y - mus)**2).mean()
```

```python
plt.scatter(X, mus)
plt.scatter(X, Y)
```
